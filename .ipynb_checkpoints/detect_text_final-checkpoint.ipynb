```python
import  cv2
import glob
#from pre_processing import preProcessing
import tkinter as tk
from tkinter import filedialog
import matplotlib.pyplot as plt
import numpy as np
sqKernel = cv2.getStructuringElement(cv2.MORPH_RECT, (16, 32))
root = tk.Tk()
root.withdraw()
    # Ask the user to select an image file using a file dialog box
file_path = filedialog.askopenfilename(filetypes=[("Image files", "*.jpg;*.jpeg;*.png")])
    # load the image, resize it, and convert it to grayscale
image = cv2.imread(file_path)
# Get the size of the image
height, width, channels = image.shape

print("Image size:")
print("Height:", height)
print("Width:", width)
print("Channels:", channels)
IMAGE_copy = image.copy()
plt.imshow(image)
plt.show()
```

```python
grayImg = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grayImg)
plt.show()
```

```python
ret, thresh = cv2.threshold(grayImg, 0, 255, cv2.THRESH_OTSU | cv2.THRESH_BINARY_INV)
plt.imshow(thresh)
plt.show()
```

```python
# Apply connected component analysis
output = cv2.connectedComponentsWithStats(thresh, connectivity=8)

# Get the statistics from the connected components analysis
num_labels = output[0]
labels = output[1]
stats = output[2]

# Define a threshold for the minimum area of objects to keep
min_area_threshold = 120  # Adjust this value as per your requirements

# Create a mask to filter out small objects
mask = np.zeros_like(thresh)
for label in range(1, num_labels):
    area = stats[label, cv2.CC_STAT_AREA]
    if area >= min_area_threshold:
        mask[labels == label] = 255

# Apply the mask to the thresholded image
filtered_thresh = cv2.bitwise_and(thresh, mask)
filtered_thresh_copy = filtered_thresh.copy()
plt.imshow(filtered_thresh)
plt.show()
```

```python
filtered_thresh = cv2.morphologyEx(filtered_thresh, cv2.MORPH_CLOSE, sqKernel)
filtered_thresh = cv2.erode(filtered_thresh, None, iterations=1)
plt.imshow(filtered_thresh)
plt.show()
```

```python
import numpy as np
import pytesseract
from tensorflow.keras.models import load_model
import imutils
import tensorflow as tf

model = load_model("resnet-az.h5")
labelNames = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Find contours of the yellow regions in the mask
contours, _ = cv2.findContours(filtered_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Sort contours left-to-right
contours = sorted(contours, key=lambda ctr: cv2.boundingRect(ctr)[0])

# Initialize output variable
output_list = []

# Perform OCR on each rectangle
for contour in contours:
    (x, y, w, h) = cv2.boundingRect(contour)
    # Draw a rectangle around the current region
    # Increase the rectangle size by a factor (e.g., 10 pixels)
    padding = 15
    cv2.rectangle(IMAGE_copy, (x - padding, y - padding), (x + w + padding, y + h + padding), (0, 255, 0), 2)
    # Adjust the padding value as needed

    # Extract character region from word image with padding
    word_region = image[max(0, y - padding):min(image.shape[0], y + h + padding),
                        max(0, x - padding):min(image.shape[1], x + w + padding)]

    if word_region.size == 0:
        # Skip empty regions
        continue

    # Convert to grayscale
    word_region = cv2.cvtColor(word_region, cv2.COLOR_BGR2GRAY)

    # Threshold the character region to obtain binary image
    _, word_thresh = cv2.threshold(word_region, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)

    # Find contours of the white regions in the binary image
    char_contours, _ = cv2.findContours(word_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Sort contours left-to-right
    char_contours = sorted(char_contours, key=lambda ctr: cv2.boundingRect(ctr)[0])
    plt.imshow(word_thresh)
    plt.show()
    output = ""
    for char_contour in char_contours:
        (char_x, char_y, char_w, char_h) = cv2.boundingRect(char_contour)
        #cv2.rectangle(word_region, (char_x, char_y), (char_x + char_w, char_y + char_h), (0, 0, 255), 3)
        
        char_roi = word_thresh[char_y:char_y+char_h, char_x:char_x+char_w]
#         plt.imshow(char_roi)
#         plt.show()
        # Resize character image to 32x32
        char_roi = cv2.resize(char_roi, (32, 32), interpolation=cv2.INTER_LINEAR)

        # Add channel dimension to character image
        char_roi = np.expand_dims(char_roi, axis=-1)

        # Normalize pixel values to range [0, 1]
        char_roi = char_roi.astype("float") / 255.0

        # Make prediction using OCR model
        with tf.device('/device:GPU:0'):
            probs = model.predict(np.array([char_roi]))
            prediction = probs.argmax(axis=1)
            label = labelNames[prediction[0]]
            output += label
            print(output)
        # Display text above the rectangle
    text_x = x - padding
    text_y = y - padding - 10  # Adjust the vertical position of the text
    #cv2.putText(IMAGE_copy, output, (text_x, text_y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)    
    output_list.append(output)

print(output_list)
plt.imshow(IMAGE_copy)
plt.show()
```
